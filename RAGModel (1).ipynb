```python
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader
from langchain_core.documents import Document  # Optional: for type hints

# Load all PDF documents in the 'books/' directory
pdf_loader = DirectoryLoader(
    path="/content",
    glob="*.pdf",
    loader_cls=PyPDFLoader
)

# Lazy loading documents to save memory
docs = pdf_loader.lazy_load()

# Loop through each document and print metadata
# for doc in docs:
#     print(doc.metadata)
# Loop through and print text content from each page
for i, doc in enumerate(docs):
    print(f"--- Page {i+1} ---")
    print(doc.page_content)
    print("\n")  # For spacing between pages
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

split_docs = splitter.split_documents(list(pdf_loader.load()))
print(len(split_docs))  # ✅ Use len() function
```

```python
# from langchain_community.embeddings import HuggingFaceEmbeddings
# from langchain_community.vectorstores import FAISS

# embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# vector_store = FAISS.from_documents(split_docs, embeddings)

from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma  # ✅ Use Chroma instead of FAISS

# Create embeddings
# embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
# embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/distiluse-base-multilingual-cased-v1")
# embeddings = HuggingFaceEmbeddings(model_name="intfloat/multilingual-e5-small")



# Create Chroma vector store with persistence
vector_db = Chroma.from_documents(
    documents=split_docs,
    embedding=embeddings,
    persist_directory="./chroma_db"  # ✅ This will save the vector DB to disk
)

# Optional: Persist the database to disk
# vector_db.persist()
print(vector_db)
```

```python
retriever = vector_db.as_retriever(search_type="similarity", search_kwargs={"k": 4})
```

```python
retriever
```

```python
retriever.invoke('অনুপমের ভাষায় সুপুরুষ কাকে বলা হয়েছে?')
```

```python
# from langchain_core.prompts import PromptTemplate
# prompt = PromptTemplate(
#     template="""
#       You are a helpful assistant.
#       Answer ONLY from the provided book context.
#       If the context is insufficient, just say you don't know.

#       {context}
#       Question: {question}
#     """,
#     input_variables = ['context', 'question']
# )

from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    template="""
      You are a helpful assistant.
      Answer ONLY from the provided book context.
      If the context is insufficient, just say you don't know.

      তুমি একজন সহায়ক সহকারী।
      কেবলমাত্র প্রদত্ত বইয়ের প্রসঙ্গ থেকেই উত্তর দাও।
      যদি প্রসঙ্গ অপর্যাপ্ত হয়, তাহলে শুধু বলো 'আমি জানি না'।

      {context}
      প্রশ্ন / Question: {question}
    """,
    input_variables=["context", "question"]
)
```

```python
# from langchain_community.chat_models import ChatOllama

# llm = ChatOllama(
#     model="llama3",      # or "mistral", "llama2", etc.
#     temperature=0.2,     # Optional
#     base_url="http://localhost:11434"  # or ngrok/remote URL if remote
# )
# === USE STANDARD LLM (NOT ChatOllama) ===
from langchain_ollama import OllamaLLM
llm = OllamaLLM(model="Qwen2.5")  # using PromptTemplate, not ChatPromptTemplate
```

```python
question          = "বিয়ের সময় কল্যাণীর প্রকৃত বয়স কত ছিল?"
retrieved_docs    = retriever.invoke(question)
```

```python
retrieved_docs
```

```python
context_text = "\n\n".join(doc.page_content for doc in retrieved_docs)
context_text
```

```python
final_prompt = prompt.invoke({"context": context_text, "question": question})
```

```python
final_prompt
```

```python
answer = llm.invoke(final_prompt)
print(answer)
```
